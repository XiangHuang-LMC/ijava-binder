# Feb 10: Nested for-loops, do-while loop, 3n+1 problem

```java
for (int i =0; i<10; i=i+1)
{
    System.out.println("i="+i);

}
```

```java
for (int i =0; i<10; i=i+1)
{
    for(int j=0; j<10; j=j+1)
    {
        System.out.println("("+i+"," +j+")");
    }

}
```

**Example**: Let i iterates through {2,4,6,8,10} and j iterates through {1,3,5,7,9}. Print out all the combinations of the value for pair (i,j)

```java
for (int i =2; i<11; i=i+2)
{
    for(int j=1; j<10; j=j+2)
    {
        System.out.println("("+i+"," +j+")");
    }

}
```

## While-loop

```java
int i=0;
while(i<10)
{
    System.out.println("i="+i);

    
}
```

```java
int i=0;
while (i< 10)
{
    int j=0;
    while(j<10)
    {
         System.out.println("("+i+"," +j+")");
        i=i+1;
         j=j+1;
    }
    
    
}
```

```java
int i=0;
while (i< 10)
{
    int j=0;
    while(j<10)
    {
         System.out.println("("+i+"," +j+")");
        
         j=j+1;
    }
    
    i=i+1;
    
}
```
